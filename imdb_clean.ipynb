```python
#create funtion to scrape each page
def scrape_page(x):
    URL = f'https://www.imdb.com/list/ls009487211/?sort=list_order,asc&st_dt=&mode=detail&page={x}'
    response = requests.get(URL)
    soup=bs(response.text)
    
    
    title = soup.findAll('h3', attrs={'class':'lister-item-header'})
    movie_list=[re.sub("\d+\.\s", "", x.text.strip().replace('\n', '')) for x in title]
    runtime = soup.findAll('span', attrs={'class':'runtime'})
    runtime_list = [x.text for x in runtime]
    rating = soup.findAll('span', attrs={'class':'certificate'})
    rating_list = [x.text for x in rating]
    genre = soup.findAll('span', attrs={'class':'genre'})
    genre_list=[re.sub("\d+\.\s", "", x.text.strip().replace('\n', ' ')) for x in genre]
    summary = soup.findAll('p', attrs={'class':''})[1:]
    summary = [re.sub("\d+\.\s", "", x.text.strip().replace('\nA', '')) for x in summary]
    summary = pd.DataFrame(summary)
    summary = summary[summary[0].str.contains("Best Picture Winner")==False]
    details_df = pd.DataFrame(movie_list).rename(columns={0:'title'})
    details_df['runtime'] = pd.DataFrame(runtime_list)
    details_df['rating'] = pd.DataFrame(rating_list)
    details_df['genre'] = pd.DataFrame(genre_list)
    details_df['summary'] = summary
    
    return details_df

scrape_page(1)
```

```python
#test page 2
scrape_page(2)
```

```python
df=pd.DataFrame()

for num in range(1, 480):
    df = pd.concat([df, scrape_page(num)])
```
