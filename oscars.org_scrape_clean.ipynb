```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

#URL for list of all oscar nominees over all years
URL = 'https://www.oscars.org/oscars/ceremonies/' # /{year}
#get response
response = requests.get(URL)
#make it into a soup
soup=bs(response.text)

#create soup for nominees
oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})

oscar_2022 = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4')] for x in oscar_soup}, orient = 'index').T.melt(var_name='category', value_name='nominee')
oscar_2022 = oscar_2022.dropna()
oscar_2022['year'] = 2022
oscar_2022
```

```python
#user-defined funtion for each year to get category, nominee, year
def scrape_year(x):
    URL = f'https://www.oscars.org/oscars/ceremonies/{x}' # /{year}

    response = requests.get(URL)
    soup=bs(response.text)

    oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})
    oscar = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4') if len(x.findAll('h4'))>0] for x in oscar_soup if len(x.findAll('h2'))>0}, orient = 'index').T.melt(var_name='category', value_name='nominee')
    oscar = oscar.dropna()
    oscar['year'] = x
    return oscar
    
scrape_year(2022)
```

```python
df = pd.DataFrame()
#loop to go through years, concating as you go 

for num in range(1929, 2023):
    df = pd.concat([df, scrape_year(num)])
```

```python
#user-defined funtion for single year with winners
def scrape_winner(x):
    URL = f'https://www.oscars.org/oscars/ceremonies/{x}' # /{year}

    response = requests.get(URL)
    soup=bs(response.text)

    winner_soup = soup.findAll('div', attrs={'class':'view-grouping'})
    winner = pd.DataFrame.from_dict({x.find('h2').text: [y.text for y in x.findAll('h4') if len(x.findAll('h4'))>0][0] for x in winner_soup if len(x.findAll('h2'))>0}, orient = 'index').reset_index().rename(columns={'index':'category', 0:'nominee'})
    winner['year'] = x
    return winner
    
scrape_winner(1929)
```

```python
#use funtion to get all winners, all years
winner_df = pd.DataFrame()
#loop to go through years, concating as you go 

for num in range(1929, 2023):
    winner_df = pd.concat([winner_df, scrape_winner(num)])
```
