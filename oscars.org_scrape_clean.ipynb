```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

URL = 'https://www.oscars.org/oscars/ceremonies/' # /{year}

response = requests.get(URL)
soup=bs(response.text)

oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})

oscar_2022 = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4')] for x in oscar_soup}, orient = 'index').T.melt(var_name='category', value_name='nominee')
oscar_2022 = oscar_2022.dropna()
oscar_2022['year'] = 2022
oscar_2022
```

```python
#user-defined funtion for each year
def scrape_year(x):
    URL = f'https://www.oscars.org/oscars/ceremonies/{x}' # /{year}

    response = requests.get(URL)
    soup=bs(response.text)

    oscar_soup = soup.findAll('div', attrs={'class':'view-grouping'})
    oscar = pd.DataFrame().from_dict({x.find('h2').text: [y.text for y in x.findAll('h4') if len(x.findAll('h4'))>0] for x in oscar_soup if len(x.findAll('h2'))>0}, orient = 'index').T.melt(var_name='category', value_name='nominee')
    oscar = oscar.dropna()
    oscar['year'] = x
    return oscar
    
scrape_year(2022)
```

```python
scrape_year(2021)
```

```python
df = pd.DataFrame()
#loop to go through years, concating as you go 

for num in range(1929, 2023):
    df = pd.concat([df, scrape_year(num)])
```
