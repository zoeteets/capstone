```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

URL = 'https://en.wikipedia.org/wiki/Academy_Award_for_Best_Picture'

response = requests.get(URL)

response.status_code
```

```python
soup = bs(response.text)
```

```python
soup.findAll('table')
```

```python
soup.find('table', attrs={'class':'wikitable'})
```

```python
table_html = str(soup.findAll('table', attrs={'class':'wikitable'}))

from IPython.core.display import HTML

HTML(table_html)
```

```python
nominees = pd.concat([twenties, thirties, fourties, fifties, sixties, seventies, eighties, nineties, thousand, tens, twotwenties], ignore_index=True)
nominees
```

```python
nominees.info()
```

```python
nominees['oscars_year']= nominees["Year of Film Release"].str.extract("(\d{4})")
nominees
```

```python
nominees.columns = map(str.lower, nominees.columns)
nominees.columns = nominees.columns.str.replace(' ', '_')
nominees.info()
```

```python
nominees['oscars_year'] = pd.to_numeric(nominees['oscars_year'])
```

```python
nominees.info()
```

```python
nominees
```

```python
BP_nominees = nominees.dropna(thresh=3)
BP_nominees = nominees.drop(columns={'year_of_film_release'})
BP_nominees.to_csv('bp_nominees.csv')
BP_nominees
```
