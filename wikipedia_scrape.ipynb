```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

URL = 'https://en.wikipedia.org/wiki/Academy_Award_for_Best_Picture'

response = requests.get(URL)

response.status_code
```

```python
soup = bs(response.text)
```

```python
#soup.findAll('table')
```

```python
soup.find('table', attrs={'class':'wikitable'})
```

```python
table_html = str(soup.findAll('table', attrs={'class':'wikitable'}))

from IPython.core.display import HTML

HTML(table_html)
```

```python
twenties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[0]
twenties = twenties.dropna(thresh=2).rename(columns={'Film Studio':'nominee'})
twenties
#nominee types vary through history. originally studios, now producers
```

```python
thirties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[1].dropna(thresh=2).rename(columns={'Film Studio':'nominee'})
thirties
```

```python
fourties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[2].dropna(thresh=2).rename(columns={'Film Studio':'nominee'})
fourties
```

```python
fifties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[3].dropna(thresh=2).rename(columns={'Film Studio/Producer(s)':'nominee'})
fifties
```

```python
sixties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[4].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
sixties
```

```python
seventies = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[5].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
seventies
```

```python
eighties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[6].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
eighties
```

```python
nineties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[7].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
nineties
```

```python
thousand = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[8].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
thousand
```

```python
tens = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[9].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
tens
```

```python
twotwenties = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[10].dropna(thresh=2).rename(columns={'Producer(s)':'nominee'})
twotwenties
```

```python
studios = pd.read_html(str(soup.findAll('table', attrs={'class':'wikitable'})))[11]
studios
```

```python
nominees = pd.concat([twenties, thirties, fourties, fifties, sixties, seventies, eighties, nineties, thousand, tens, twotwenties], ignore_index=True)
nominees = nominees.drop(columns={'Films'})
nominees.head()
```

```python
nominees.info()
```

```python
nominees.tail()
```

```python
nominees['release_year']= nominees["Year of Film Release"].str.extract("(\d{4})")
nominees.tail()
```

```python
#drop extra year column..ALERT- must change "nominee" to different name to avoid future confusion
nominees = nominees.drop(columns={'Year of Film Release'}).rename(columns={'nominee':'producer'})
nominees.head()
```

```python
nominees.columns = map(str.lower, nominees.columns)
nominees.info()
```

```python
nominees['release_year'] = pd.to_numeric(nominees['release_year'])
```
