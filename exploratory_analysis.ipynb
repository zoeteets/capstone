```python
import pandas as pd
```

```python
# manually edited errors in Excel
# bring edited csv back in
bp_full_df = pd.read_csv('bp_full_df2.csv')
bp_full_df = bp_full_df.drop(columns={'Unnamed: 0'})
bp_full_df
```

```python
# split genre column into multiples
genres = bp_full_df['genre'].str.split(pat =',', expand=True)
genres
```

```python
genres = genres.rename(columns={0:'genre_1', 1:'genre_2', 2:'genre_3'})
genres
```

```python
#add genre df to bp_full_df based on index
bp_full_df = pd.concat([bp_full_df, genres], axis=1)
bp_full_df
```

```python
#drop unneccessary columns
bp_full_df = bp_full_df.drop(columns={'matches'})
```

```python
bp_full_df.head()
```

```python
#clean runtime, int only
bp_full_df['runtime_min'] = bp_full_df['runtime'].str.extract('(.*)\s')
bp_full_df = bp_full_df.drop(columns={'runtime'})
bp_full_df['runtime_min'] = pd.to_numeric(bp_full_df['runtime_min'])
#verify
bp_full_df.info()
```

```python
bp_full_df['gross'] = bp_full_df['gross'].str.replace(',', '')
bp_full_df['gross'] = bp_full_df['gross'].str.replace('\.00', '')
bp_full_df
```

```python
bp_full_df['gross'] = bp_full_df['gross'].astype('Int32')
bp_full_df.head()
```

```python
#save new df as csv to import to SQL
bp_full_df.to_csv('bp_clean.csv', index=False)
```

```python
winners = bp_full_df.query('winner == 1')
winners.head()
```

```python
losers = bp_full_df.query('winner == 0')
losers.head()
```

```python
#genre most won - R
perc_genre_won = pd.DataFrame(winners['rating'].value_counts()).reset_index().rename(columns={'rating':'count', 'index':'rating'})
perc_genre_won['total_won'] = winners['title'].count()
perc_genre_won['percent'] = perc_genre_won['count']/perc_genre_won['total_won']*100
perc_genre_won
```

```python
#genre most lost - R
perc_genre_lost = pd.DataFrame(losers['rating'].value_counts()).reset_index().rename(columns={'rating':'count', 'index':'rating'})
perc_genre_lost['total_lost'] = losers['title'].count()
perc_genre_lost['percent'] = perc_genre_lost['count']/perc_genre_lost['total_lost']*100
perc_genre_lost
```

```python
#Average runtime of winners is longer than avg runtime of losers
print('winner_avg_runtime:')
print(winners['runtime_min'].mean())
print('loser_avg_runtime:')
print(losers['runtime_min'].mean())
```
